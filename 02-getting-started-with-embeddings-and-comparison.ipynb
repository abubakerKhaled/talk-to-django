```python
import numpy as np
from sentence_transformers import SentenceTransformer
```

```python
# 1. Load a pretrained Sentence Transformer model
model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
## this is the index representation of the words we want the number representation
split_docs = sorted(["The dog jumped over the cat"][0].split(' '))
split_docs
```

```python
docs = ["The dog jumped over the cat"]
docs
```

```python
## this way we make a number representaion of the list docs that is numpy array
embeddings = model.encode(docs, clean_up_tokenization_spaces=False)
embeddings
```

```python
docs2 = ["The cat jumped over the dog"]
```

```python
## this looks the same as the split_docs in the first 
split_docs2 = sorted(docs2[0].split(' '))
split_docs2
```

```python
## but in this it looks very different from the first embeddings so the embeddings
## is perfect for comparison things...
embeddings2 = model.encode(docs2, clean_up_tokenization_spaces=False)
embeddings2
```

```python
## we see here that the two embeddings are similar but 
model.similarity(embeddings2, embeddings2)
```

```python
## in this we see the similarity is slite different
model.similarity(embeddings, embeddings2)
```
